```python
%matplotlib inline
import matplotlib.pyplot as plt
from mwviews.api import PageviewsClient
import pandas as pd
```

```python
p = PageviewsClient()
```

```python
project = 'en'
articles = ['RSA_Factoring_Challenge', 'P_versus_NP_problem', 'Millennium_Prize_Problems', 'BQP']
st = '20170101'
end = '20170814'
```

```python
resp = p.article_views(project + '.wikipedia', articles, start=st, end=end)
```

```python
df = pd.DataFrame(resp).transpose()
```

```python
df.columns
```

```python
plt.figure(figsize=(15,5))
plt.plot(df)
plt.legend(df.columns)
plt.show()
```
